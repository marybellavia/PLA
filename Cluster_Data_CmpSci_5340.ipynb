```python
import pandas as pd
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
#creating linearly separable training data
x, y = make_blobs(n_samples=50, centers=2, n_features=2, cluster_std= 1,
                  random_state=8)
#outputs 2 clusters, with the y output corresponding to a given cluster. In the case of 2 clusters a y=0 value will mark the first cluster, and a y=1 will mark the cluster of the second value. 
#this can be scaled up for n clusters
fig = plt.figure(figsize=(15,12))
#Arbitrary sizing, I just picked it because I like having a big plot
plt.plot(x[:, 0][y == 0], x[:, 1][y == 0], 'gs')
plt.plot(x[:, 0][y == 1], x[:, 1][y == 1], 'bs')
#These plot all ordered pairs where the "y" value of zero defined in the "make_blobs" function corresponds to the first cluster with a green square.
#The "y" value defined in the second plot command orresponds to a blue square.
#For the random_state and cluster STD the values are linearly seperable
plt.xlabel("X")
plt.ylabel("Y")
plt.title('Linearly Separable')
```

```python
# creating NON linearly separable training data
x2, y2 = make_blobs(n_samples=50, centers=2, n_features=2, cluster_std= 3,
                  random_state=8)
#For this section, we can make a non lineraly seperable sample by scaling up the cluster_std 
fig = plt.figure(figsize=(15,12))
plt.plot(x2[:, 0][y2 == 0], x2[:, 1][y2 == 0], 'gs')
plt.plot(x2[:, 0][y2 == 1], x2[:, 1][y2 == 1], 'bs')
#These plot all ordered pairs where the "y" value of zero defined in the "make_blobs" function corresponds to the first cluster with a green square.
#The "y" value defined in the second plot command orresponds to a blue square.
#For the random_state and cluster STD the values are linearly seperable
plt.xlabel("X")
plt.ylabel("Y")
plt.title('Not Linearly Separable')
```
